```python
    # Exercises
    # 1 Create a tensor a from list(range(9)). Predict and then check the size, offset,
    # and stride.
    # a Create a new tensor using b = a.view(3, 3). What does view do? Check
    # that a and b share the same storage.
    # b Create a tensor c = b[1:,1:]. Predict and then check the size, offset, and
    # stride.

import torch

t = torch.tensor(list(range(9)))
a = t
# size is 9, offset is 0, stride is 1
print("Tensor a:", t)

b = t.view(3, 3) # view does not change the storage, it just changes the shape

print("b stride:", b.stride())
print("Tensor b (view of a):", b)
print("Storage of a and b are the same:", t.storage().data_ptr() == b.storage().data_ptr())

a.storage().data_ptr()== b.storage().data_ptr()
```

```python
c = b[1:,1:] # c size is 2*2, offset is 4, stride is (3,1)
c.storage_offset()
```
